```python
from matplotlib.animation import FuncAnimation
from matplotlib.animation import FFMpegWriter
```

```python
fig = plt.figure()
def animation(frame) :
    fig.clear()
    t = frame
    state_f = f.state_t(state_i, t , M)
    sol = f.solution(state_f, mN, pN)
    plt.title(f"t= {t}")
    plt.xlabel("num of protein")
    plt.ylabel("num of mRNA")
    plt.imshow(sol, origin="lower")
    plt.colorbar()
    
writer=FFMpegWriter(bitrate=500)
ani = FuncAnimation(fig, animation, frames = np.linspace(0,1,101))
ani.save('solution.gif', writer='imagemagick')
#ani.save("Solution.gif", dpi=300, writer=writer)    
```
